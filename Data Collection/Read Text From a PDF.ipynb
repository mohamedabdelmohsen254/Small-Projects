```python
# Import the PdfReader class from the PyPDF2 module
from PyPDF2 import PdfReader

# Create a PdfReader object for the specified pdf file
reader = PdfReader("Aurélien Géron - Hands-On Machine Learning with Scikit-Learn, Keras, and TensorFlow  Concepts, Tools, and Techniques to Build Intelligent Systems-O'Reilly Media, Inc. (2023) 3rd Edition.pdf")

# Get the number of pages in the pdf
number_of_pages = len(reader.pages)

# Print the number of pages
print(number_of_pages)
```

```python
# Get the 20th page of the pdf
page = reader.pages[20]

# Extract the text from the page
text = page.extract_text()

# Print the extracted text
text
```

```python
# Initialize an empty list to store the text data
text_data=[]

# Loop through the first 10 pages of the pdf
for i in range(0,10):
    # Extract the text from the current page
    pages = reader.pages[i].extract_text()
    # Append the extracted text to the text_data list
    text_data.append(pages)

# Print the text_data list
text_data  
```

## Scraping From a Website

```python
import requests

from bs4 import BeautifulSoup


# 2nd step use requests to fetch the url
result = requests.get("https://en.wikipedia.org/wiki/Machine_learning")
src = result.content

# 4th step create soup object to parse content
soup = BeautifulSoup(src, "html") # we can use lxml parser
#print(soup)
text = soup.findAll("p")
```
