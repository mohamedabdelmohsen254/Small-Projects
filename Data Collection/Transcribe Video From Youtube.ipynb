This code uses the YouTube module and its function "YouTube()" to access the video from a given url, videourl. Then it filters the streams of the video to get the progressive download, mp4 format and the highest resolution stream. It then checks if the specified path for the downloaded video, path, exists on the local system, and if not, it creates the directory. Finally, it uses the download() function to download the video to the specified path on the local system.

```python
from pytube import YouTube
import os

def downloadYouTube(videourl, path):
    # Initialize YouTube object with the video url
    yt = YouTube(videourl)
    # Filter the video streams to get the progressive download, mp4 format and the highest resolution stream.
    yt = yt.streams.filter(progressive=True, file_extension='mp4').order_by('resolution').desc().first()
    # Check if the specified path for the downloaded video exists on the local system, if not, create the directory.
    if not os.path.exists(path):
        os.makedirs(path)
    # Download the video to the specified path on the local system.
    yt.download(path)
```

### convert audio to text

The record() method is used to record the audio from the source and 
The recognize_google() method transcribes the audio and specifies the language as English. Finally, it prints the transcribed text.

```python
import speech_recognition as sr

# Initialize the Recognizer class
r = sr.Recognizer()

# Load the audio file
audio_file = sr.AudioFile('/content/Machine Learning.wav')

# Open the audio file using a context manager
with audio_file as source:
    # Record the audio
    audio = r.record(source)

# Use the recognize_google() method to transcribe the audio and specify the language
text = r.recognize_google(audio, language='en')

# Print the transcribed text
print(text)
```

The error is due to the size of the video is large so, I will segment it

```python
#Importing library and thir function
from pydub import AudioSegment
from pydub.silence import split_on_silence

#reading from audio mp3 file
sound = AudioSegment.from_mp3("/content/Machine Learning.wav")

# spliting audio files
# split on silences longer than 1000ms 
# anything under -16 dBFS is considered silence
audio_chunks = split_on_silence(sound, min_silence_len=1000, silence_thresh=-40 )

```

```python
#print the size of audio chunks
len(audio_chunks)
```

Now we are ready to convert the audio to text

```python
text=""
for chunk in audio_chunks:
    # Create the output file
    output_file = "/content/Machine Learning.wav"
    # Export the current chunk to a wav file
    chunk.export(output_file, format="wav")

    # Load the audio file
    audio_file = sr.AudioFile(output_file)

    # Open the audio file using a context manager
    with audio_file as source:
        # Record the audio
        audio = r.record(source)
    try:
        # Use the recognize_google() method to transcribe the audio and specify the language
        text += " "+r.recognize_google(audio, language='en')
    except:
        # In case of any error, continue to the next chunk
        continue
```
