```python
from bs4 import BeautifulSoup 
import pandas as pd
import requests
import csv 
import re
```

```python
def date():
    # Prompt the user to enter a date in the format MM/DD/YYYY
    date = input("Please enter a date in the following formar MM/DD/YYYY:")
    
    # create a regular expression pattern to match the date format
    date_pattern = re.compile(r'^(0[1-9]|1[012])[/](0[1-9]|[12][0-9]|3[01])[/](19|20)\d\d$')
    
    # check if the entered date matches the pattern
    if not date_pattern.match(date):
        raise ValueError("Invalid date format. Please enter date in the format MM/DD/YYYY")
    
    # Use the input date to create a request to YallaKora website using the python requests library
    
    try:
        # send a get request to the website using the input date
        page = requests.get(f"https://www.yallakora.com/match-center/%D9%85%D8%B1%D9%83%D8%B2-%D8%A7%D9%84%D9%85%D8%A8%D8%A7%D8%B1%D9%8A%D8%A7%D8%AA?date={date}")
    
    except requests.exceptions.RequestException as e:
        # raise an exception if there is any error connecting to the website
        raise ValueError("Error connecting to the website: ", e)
        
    # Return the page object returned by the website
    return page
```

```python
def main():
    page = date()
    
    # Initialize an empty list to save the details of each match
    matches_details = []
    
    # Get the source code of the page 
    src= page.content 
    
    # Create a BeautifulSoup object to parse the source code
    soup = BeautifulSoup(src, "lxml") 
    
    #get world cup matches only
    championships = soup.find_all("div", {'class':'matchCard'})
    
    # Define a function to extract the match information
    def get_match_info(championships):
        # Get the title of the championship
        championship_title = championships.contents[1].find('h2').text.strip()
        #championship_title = championships.find("div",{'class':'title'}).find('h2').text.strip()
        
        # Find all the matches in the championship
        all_matches= championships.contents[3].find_all('li')
        
        # Get the number of matches
        num_of_matches = len(all_matches)
        
        # Iterate through each match
        for i in range(num_of_matches):
            # Get the names of the teams
            team_A=all_matches[i].find('div',{'class':'teams teamA'}).text.strip()
            team_B=all_matches[i].find('div',{'class':'teams teamB'}).text.strip()
            
            # Get the score of the match
            match_result = all_matches[i].find('div',{'class':'MResult'}).find_all("span",{'class':'score'})
            score = f"{match_result[0].text.strip()} - {match_result[1].text.strip()}"
           
            # Get the time of the match
            match_time = all_matches[i].find('div',{'class':'MResult'}).find("span",{'class':'time'}).text.strip()
            
            #add match info to matches_details
            matches_details.append({'نوع البطوله':championship_title, 'الفريق الاول':team_A, 'الفريق الثاني':team_B,
                                   "ميعاد المباراه":match_time, "النتيجه":score})
    
    # Iterate through each championship
    for i in range(len(championships)):
        # Call the function to extract the match information for each championship
        get_match_info(championships[i])
        
    # Convert the matches_details list to a dataframe
    df =pd.DataFrame(matches_details)
    
    #Return the dataframe
    return df


        
```

```python
main()
```
